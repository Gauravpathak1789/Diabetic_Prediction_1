```python
import json
class Nutrition_Calculator:  #HERE CLASS is created name Nutrition Calculator
   
        
        def __init__(self): #Here I have implemented constructor.
            data={}
            
            self.name=input("enter the name of child-  ") #child's name taken as inpput
            data["name"]=self.name
        
            self.age=int(input("enter the age of child in year between 1 to 8 - ")) # child's age taken as input
            data["age"]=self.age
        
            self.gender=input("enter the gender of the child [Male/Female]-  ") # child's gender as input
            data["gender"]=self.gender
        
            self.height=float(input("enter the height of child in inches-  ")) # child's height as input
            data["height"]=self.height
        
            self.weight=float(input("enter the weight of child in pounds-  ")) # child's weight as input
            data["weight"]=self.weight
        
            with open("dic.json",'a') as f: #here the entered data such as name,age gender etc. is 
                                            #going to stored in json file name dic and it is in append mode
                json.dump(data,f)
        
            self.Calculator_menu() # Here I have called menu method
        
        
        def Calculator_menu(self): #Here it is menu method 
            user_input=input("""Proceed further as per your choice--
            1.BMI Calculation
            2.Daily Calorie Requirement
        """)
        
            if user_input=="1":
                self.BMI_Calculation()
            
            elif user_input=="2":
                self.Daily_Calorie_intake_Calculation()
            
            else:
                print("Wrong input ,try again")
            
            
        def BMI_Calculation(self): #BMI calculation method
            
            BMI = (self.weight/(self.height**2))*703
            
            if (BMI<16):
                print("Severely Underweight")
        
            elif (BMI>=16 and BMI<18.5):
                print("Underweight")
            
            elif (BMI>=18.5 and BMI<25):
                print("Healthy")
            
            elif (BMI>=25 and BMI<30):
                print("Overweight")
            
            elif (BMI>=30):
                print("Obese")
            
            
      
        def Daily_Calorie_intake_Calculation(self): #here calorie intaken by child is calculated
            num_food_items=int(input("number of food items taken:-  "))
        
            Food_intake_list={}
            self.Total_calorie_intake=0
        
            for i in range(num_food_items):
                
                Food_item_intake = input("enter the name of food:-  ")
            
                Quantity = int(input("enter the quantity of food intake in grams:-  "))
            
                Food_intake_list[Food_item_intake]=Quantity
            
             
        
            for i in Food_intake_list:
            
                if i=="Milk":
                    self.Total_calorie_intake=self.Total_calorie_intake+(100/100)*Food_intake_list[i]
                
                elif i=="Egg":
                    self.Total_calorie_intake=self.Total_calorie_intake+(155/100)*Food_intake_list[i]
        
                elif i=="Rice":
                    self.Total_calorie_intake=self.Total_calorie_intake+(130/100)*Food_intake_list[i]
        
                elif i=="Lentils":
                    self.Total_calorie_intake=self.Total_calorie_intake+(113/100)*Food_intake_list[i]
        
                elif i=="Vegetables":
                    self.Total_calorie_intake=self.Total_calorie_intake+(85/100)*Food_intake_list[i]
    
                elif i=="Meat":
                    self.Total_calorie_intake=self.Total_calorie_intake+(143/100)*Food_intake_list[i]
                
                else:
                    print("wrong input or empty food list")
                
            self.Min_Calorie_Requirement()  #here Min_Calorie_Requirement method is called
            
            
        def Min_Calorie_Requirement(self): #this method provide that child is nourished or not according to daily calorie requirement
            print("Child is Undernourished or not? --")
        
            if (self.age>=0 and self.age<2) and (self.Total_calorie_intake>=800 and self.Total_calorie_intake<1400):
                print("Nourished")
            
            elif (self.age>=2 and self.age<4) and (self.Total_calorie_intake>=1400 and self.Total_calorie_intake<1800):
                print("Nourished")
            
            elif (self.age>=4 and self.age<8) and (self.Total_calorie_intake>=1800 and self.Total_calorie_intake<2200):
                print("Nourished")
            
            else:
                print("Unourished")
            
            
        
S1=Nutrition_Calculator() #here we can make different objects as diiferent child
                        #for which we can calculate nutrition 
        
   
```
